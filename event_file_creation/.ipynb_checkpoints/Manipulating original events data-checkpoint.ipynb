## Manipulating original data:

The idea here is to take the demo events data, sample activity of random users from it, then add features fertilizer, pesticide, rabi,kharif, to it 

````python
import json
import numpy as np
import pandas as pd
import random
import gzip
import jsonlines
````

````python
data = [json.loads(line) for line in open('D:/Samagra/metarank_testing/data/events.jsonl/events.jsonl', 'r',encoding="utf8")]
````

````python
##checking unique vents 
event = [x["event"] for x in data ]
np.unique(np.array(event))

input_dict = data.copy()
metadata_dict = [x for x in input_dict if x["event"] == "metadata"]

input_dict = data.copy()
ranking_dict = [x for x in input_dict if x["event"] == "ranking"]

input_dict = data.copy()
interaction_dict = [x for x in input_dict if x["event"] == "interaction"]
len(metadata_dict)
len(ranking_dict)
len(interaction_dict)

##checking 10 users
users = [x["user"] for x in ranking_dict ]
users =  np.unique(np.array(users))
users_10 =  users[0:10]

##filtering for rankings exposed to those 10 users 
input_dict = ranking_dict.copy()
ranking_dict_10 =  [x for x in input_dict if x["user"] in users_10]
len(ranking_dict_10)


## filtering for interctions for those 10 users : 
input_dict = interaction_dict.copy()
interaction_dict_10 =  [x for x in input_dict if x["user"] in users_10]
len(interaction_dict_10)

## checking for items exposed in the rankings for the interactions of these 10 users 
items_10 = [x["items"] for x in ranking_dict_10 ]

item_array_10 = np.array([])

item_array_10 = np.array([])

for x in items_10:
    for ii in range(0,len(x)):
        item_array_10 = np.append(item_array_10 ,x[ii]['id'] )
## filtering metadata for those 10 items only :     
input_dict = metadata_dict.copy()
metadata_dict_10 =  [x for x in input_dict if x["item"] in item_array_10]
len(metadata_dict_10)


for x in metadata_dict_10:
    x["event"]='item'
````

````python
interaction_pd = pd.DataFrame(list(range(0,272)))
````

### checking interactions and filtering interactions of 4 random users to assign them "kharif" values

````python
interaction_pd['user_id']='user'
interaction_pd['item_id']='item'

index= 0
for x in interaction_dict_10:
    interaction_pd.loc[index,'user_id'] = x['user']
    interaction_pd.loc[index,'item_id'] = x['item']
    index = index + 1
````

````python
A_filter_users =np.array(['0014aa0e8f2c81543737bbbf69e7e262','002c41d48be25cfafbd3a447539834c3','006b48674b9bf1ced7bed88aa26cb78e','013a5348467a3d48e1a24f4def011abe'])
users_10_A =  users_10[np.isin(users_10,A_filter_users)]
users_10_B = users_10[~np.isin(users_10,A_filter_users)]
````

````python
interaction_pd['User_type'] =  'B'
interaction_pd.loc[np.isin(interaction_pd.user_id,users_10_A),'User_type']= "A"
````

````python
df1 = interaction_pd[['item_id','User_type']].drop_duplicates().reset_index()
````

````python
items_A = df1.item_id[df1.User_type=="A"]
df1["User_type_2"] = "B"
df1.loc[np.isin(df1.item_id , items_A),"User_type_2"]= "A"
items_A_array = items_A.values
````

````python
for x in metadata_dict_10:
    if x["item"] in items_A_array:
        fertilizer = random.choice(['fertilizer_2','fertilizer_3'])
        pesticide = 'pesticide_1'
    else:
        fertilizer = 'fertilizer_1'
        pesticide = random.choice(['pesticide_2','pesticide_3'])
    
    x["fields"].extend([{
                "name": "pesticide",
                "value": fertilizer
                },
            {
                "name": "fertilizer",
                "value": pesticide
                }])
````

````python
no_of_users = 10
````

````python
user_dic1 = []

def condition_crop(id1):
    if users_10[id1] in users_10_A:
        crop = 'kharif'
    else:
        crop = 'rabi'
    return(crop)


for id1 in range(no_of_users):
   
    user_dic1.append({
        "event": "user",
        "id": str(id1 + 1),
        "timestamp":  metadata_dict_10[0]['timestamp'],
        "user": users_10[id1],
        "fields": [{
                "name": "land_area",
                "value": random.randint(100,250)/100.00
                },
            {
                "name": "crops",
                "value": condition_crop(id1)
                }]
    })

user_dic1[4:6]
````

````python
data_10_list = list()

data_10_list.extend(metadata_dict_10)
data_10_list.extend(ranking_dict_10)
data_10_list.extend(interaction_dict_10)
data_10_list.extend(user_dic1)

filename =  "data/events_small_kharif_rabi.jsonl.gzip"
payload = data_10_list.copy()
````

````python
def writeall_jsonl_gz(filename, payload ):
    with gzip.open(filename, 'wb') as fp:
        json_writer = jsonlines.Writer(fp)
        json_writer.write_all(payload)
````

````python
writeall_jsonl_gz(filename, payload )
````
